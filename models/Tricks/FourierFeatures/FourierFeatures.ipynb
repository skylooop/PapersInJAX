```python
import torch
import torch.nn as nn
```

```python
from PIL import Image
import numpy as np

flower = np.asarray(Image.open("flower.jpeg").resize((224, 224))) / 255.
flower.shape
```

```python
device = "cuda"
flower = torch.tensor(flower).unsqueeze(0).permute(0, 3, 2, 1).to(device)
```

```python
import matplotlib.pyplot as plt

def show_tensor(x):
    return x.permute(1, 2, 0).cpu().numpy()

plt.imshow(show_tensor(flower[0]))
```

```python
coords = np.linspace(0, 1, flower.shape[-1], endpoint=False)
xy = np.stack(np.meshgrid(coords, coords), -1)
xy = torch.tensor(xy).unsqueeze(0).permute(0, 3, 1, 2).float().contiguous().to(device)
```
